```python
import numpy as np
import pandas as pd

df_can = pd.read_excel('https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true', 
                       sheet_name = 'Canada by Citizenship',skiprows=range(20),skipfooter=2)

print('Data read into a pandas dataframe!')
```

```python
print(df_can.head())
print(df_can.tail())
print(df_can.info())
#ketika kita show info datanya terlihat tidak ada yang null
print(df_can.columns.values)
print(df_can.index.values)
print(df_can.shape)
```

```python
#pembersihan data

#df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace = True)
df_can.head(2)
df_can.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
print(df_can.columns)

print(df_can.isnull().sum())
print(df_can.describe())
```

```python
print(df_can.Country)
print(df_can[['Country',1980,1981,1982,1983,1984,1985]]) #Seperti Select [nama_kolom1, nama_kolom_2 dst]

#1 the full row data
print(df_can.set_index('Country',inplace = True))
df_can.index.name=None
print(df_can.loc['Japan'])
print(df_can.iloc[87])
print(df_can[df_can.index=='Japan'].T.squeeze())
```

```python
#2. for year 2013

print(df_can.loc['Japan',2013])
print(df_can.iloc[87,36])

#3 for years 1980 to 1985
print(df_can.loc['Japan',[1980,1981,1983,1984,1985]])
print(df_can.iloc[87,[3,4,5,6,7,8]])
df_can.columns = list(map(str,df_can.columns))

years = list(map(str,range(1980,2014)))
print(years)
```

```python
#1 create the condition boolean series 
condition = df_can['Continent']=='Asia'
print(condition)

#2 pass this condition into the dataFrame
df_can[condition]
```

```python
df_can[(df_can['Continent']=='Asia')&(df_can['Region']=='South-Eastern Asia')]
```

```python
print('data dimensions:',df_can.shape)
print(df_can.columns)
df_can.head(2)
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

filipina = df_can.loc['Philippines',years]
print(filipina.head())

indonesia = df_can.loc['Indonesia',years]
print(indonesia.head())

indonesia.plot()
filipina.plot()

#indonesia.index = indonesia.index.map(int)
indonesia.plot(kind='line')
plt.title('Immigration from SEA')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

#diagram garis bisa dicoba sendiri ya..
```

```python
#Area Plots
df_can.sort_values(
    ['Total'], 
    ascending=False, axis=0, inplace=True)

```

```python
df_top5= df_can.head()
df_top5 = df_top5[years].transpose()
df_top5.head()

```

```python
df_can[(df_can['Continent']=='Asia')&(df_can['Region']=='South-Eastern Asia')]
```

```python

df_sea =df_can[(df_can['Continent']=='Asia')&(df_can['Region']=='South-Eastern Asia')]
df_sea['Total'] = df_sea.sum()
df_sea.sort_values(
    ['Total'], 
    ascending=False, axis=0, inplace=True)

df_top5sea = df_sea.head()

df_top5sea = df_top5sea[years].transpose()

df_top5sea.head()
```

```python
df_top5sea.index = df_top5sea.index.map(int)

df_top5.plot(kind='area',
             stacked=False,
             figsize=(20, 10),
             
             )

plt.title('immigration Trend of top 5 countries SEA')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_eur =df_can[(df_can['Continent']=='Europe')]
df_eur['Total'] = df_eur.sum()
df_eur.sort_values(
    ['Total'], 
    ascending=False, axis=0, inplace=True)

df_top5eur = df_eur.head()

df_top5eur = df_top5eur[years].transpose()

df_top5eur.head()
```

```python
df_top5eur.index = df_top5eur.index.map(int)

df_top5eur.plot(kind='area',
             figsize=(8,5),
             stacked = False
             
             )

plt.title('immigration Trend of top 5 countries Europe')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```
