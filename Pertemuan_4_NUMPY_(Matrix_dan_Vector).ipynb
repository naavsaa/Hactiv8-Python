```python
#Matrix Indexing

#first dimension
kelas_062 = npy.asarray([6,1,2,4,3,0,0,0])
kelas_063 = npy.asarray([1,2,3,5,6,0,0,0])
kelas_064 = npy.asarray([2,3,4,5,0,0,0,0])

#slicing

print(f"Satu dimensi Full team = {kelas_062}")

print(f"Slicing 1 : Absen pertama  / index 0 -> {kelas_062[0]}") 
print(f"Slicing 2 : Absen ketiga   / index 2 -> {kelas_062[2]}") 
print(f"Slicing 3 : Absen kelima   / index 4 -> {kelas_062[4]}") 
print("")

#second dimension

kelas_gabungan = npy.array([[6,1,2,4,3,0,0,0],
                            [1,2,3,5,6,0,0,0],
                            [2,3,4,5,0,0,0,0]])

#ini belum bisa digunakan karena ini adalah array dalam array
kelas_gabungan_array = npy.array([[kelas_062],
                            [kelas_063],
                            [kelas_064]])

#slicing

print(f"Dua dimensi Full team = {kelas_gabungan}")

print(f"""Slicing 1 : kelas pertama                 / index 0                 -> {kelas_gabungan[0]}""") 
print(f"""Slicing 2 : 2 kelas setelah pertama       / index 0 sampai 1 (2x)   -> {kelas_gabungan[0:2]}""") 
print(f"""Slicing 3 : kelas pertama, Absen kelima   / index 0,4               -> {kelas_gabungan[0,4]}""") 
print(f"""Slicing 4 : kelas kedua, Absen kelima     / index 0,4               -> {kelas_gabungan[1][4]}
      
      """) 

#third dimension

kelas_gabungan_lt2 = kelas_gabungan
kelas_gabungan_lt3 = kelas_gabungan
kelas_gabungan_lt4 = kelas_gabungan

kelas_gabungan_4_lantai = npy.array(
                            [  
                            [[6,1,2,4,3,0,0,0],
                            [1,2,3,5,6,0,0,0],
                            [2,3,4,5,0,0,0,0]],
                            
                            [[6,1,2,4,3,0,0,0],
                            [1,2,3,5,6,0,0,0],
                            [2,3,4,5,0,0,0,0]],
                            
                            [[6,1,2,4,3,0,0,0],
                            [1,2,3,5,6,0,0,0],
                            [2,3,4,5,0,0,0,0]]
                            ]
                            )

print(f"Tiga Dimensi Full team = {kelas_gabungan_4_lantai}")

print(f"""Slicing 1 : lantai pertama                               / index 0   -> 
          {kelas_gabungan_4_lantai[0]}
      
      """) 
print(f"""Slicing 2 : lantai ketiga, kelas ke 2                    / index 2   -> 
          {kelas_gabungan_4_lantai[2,1]}
      
      """) 
print(f"""Slicing 3 : lantai pertama, kelas pertama, Absen kelima  / index 0,4 -> 
          {kelas_gabungan_4_lantai[0,0,4]}
      
      """) 

```
