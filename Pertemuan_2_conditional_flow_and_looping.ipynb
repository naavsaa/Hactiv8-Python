```python
# nested if

first_sentence = 'Hello world all in python'
second_sentence = 'auf wiedersehen'
third_sentence = 'aulia bla bla bla'
fourth_sentence = 'xcx bla bla bla bla'

keyword = 'not'

#not example : the hide and seek anecdote : 

print(f"ann  : is '{keyword}' in '{first_sentence}' ?")

if keyword in first_sentence :
  print(f"boby : yes '{keyword}' is in '{first_sentence}'")
  print(f"ann  : aul is found")

elif keyword in second_sentence :
  print(f"boby : yes '{keyword}' is in {second_sentence}")
  print(f"ann  : aul is found")

elif keyword in third_sentence :
  print(f"boby : yes '{keyword}' is in '{third_sentence}'")
  print(f"ann  : aul is found")

elif keyword in fourth_sentence :
  print(f"boby : yes '{keyword}' is in {fourth_sentence}")
  print(f"ann  : aul is found")

else :
  print(f"ann  : so '{keyword}' is not found ? so not is not in english ?")

#end 

keyword = 'xcx'

#xcx example : the hide and seek anecdote : 

print(f"ann  : is '{keyword}' in '{first_sentence}' ?")

if keyword in first_sentence :
  print(f"boby : yes '{keyword}' is in '{first_sentence}'")
  print(f"ann  : aul is found")

elif keyword in second_sentence :
  print(f"boby : yes '{keyword}' is in {second_sentence}")
  print(f"ann  : aul is found")

elif keyword in third_sentence :
  print(f"boby : yes '{keyword}' is in '{third_sentence}'")
  print(f"ann  : aul is found")

elif keyword in fourth_sentence :
  print(f"boby : yes '{keyword}' is in {fourth_sentence}")
  print(f"ann  : aul is found")

else :
  print(f"ann  : so '{keyword}' is not found ? so not is not in english ?")

#try it yourself :)
```

```python
#custom use for (homework learning)

keyword = 'not'

def conversation_generator(person_1, person_2, keyword,sentence):
  print(f"{person_1}  : is '{keyword}' in '{sentence}' ?")
  if keyword in sentence :
    print(f"{person_1} : yes '{keyword}' is in '{sentence}'")
    print(f"{person_2}  : {keyword} is found")
  else :
    print(f"ann  : so '{keyword}' is not found ? so not is not in english ?")


conversation_ann_and_boby = conversation_generator

sentences = ['Hello world all in python',
             'auf wiedersehen',
             'aulia bla bla bla',
             'xcx bla bla bla bla']

keywords  = ['not',
             'bla',
             'auf',
             'xcx']

for current_sentence in sentences :
  for current_keyword in keywords :
    conversation_ann_and_boby("ann","bobby",current_keyword,current_sentence)
```

```python
dictionary = {'foo':1,'bar':2,'baz':3}
# 'k' konteksnya adalah : 
# variable temporary untuk menampung tiap data yang ada di dict 'd'
# k = key (mungkin ya di kepala yang bikin kode.id, asusmsi :p)

for k in dictionary: 
    print(k)

for k in dictionary: 
    print(dictionary[k])

for k in dictionary.values():
  print(k)

for k,v in dictionary.items():
  print(k,":",v)
```
