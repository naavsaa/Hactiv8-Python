```python
#argument : null (not required)

def calling_hacktiv8_students():
  print(f"Hi All welcome to H8 session 3")

#argument : alas cm, tinggi cm (required)

def luas_segitiga_dalam_m2(alas_cm, tinggi_cm):
  current_alas = alas_cm / 100
  current_tinggi = tinggi_cm / 100

  area = current_alas * current_tinggi / 2
  print(f"Luas segitiga adalah {area} m2")
  
  # return is to define end point of a function

  # return decimal value
  return area

  # if we want to return a "void function" we can return without area.
  # void function adalah function yang return suatu value
  # return

#Calling a function in main
calling_hacktiv8_students()

area = luas_segitiga_dalam_m2(100,100)

#jika void function maka area = none
print(area)

#luas_segitiga_dalam_m2()
```

```python
#pass by reference (penambahan dengan menggunakan rujukan(mylist))
#argument : mylist

#void function
def change_me(mylist):
  "This changes a passed list into this function"
  print("Values before the function  : ", mylist)
  
  mylist.append([1,2,3,4]);
  
  return

#not void function
def change_me_2ndtime(mylist):
  "This changes a passed list into this function"
  print("Values before the function  : ", mylist)
  
  mylist.append([0.1,0.2,0.3,0.4]);
  
  return mylist

#not void function
def change_me_3rdtime(mylist):
  "This changes a passed list into this function"
  print("Values before the function  : ", mylist)
  
  mylist.append([4,5,6,7]);
  
  return mylist

mylist = [10,20,30]
var = change_me(mylist)

print("Values after the function   : ",mylist)

#void function return none
print(var)

mylist = change_me_2ndtime(mylist)
mylist = change_me_3rdtime(mylist)

#not void function return some values
print(mylist)
```

```python
#pass by value (mau mengganti value tersebut (mylist)) 
#argument : mylist

def change_me(mylist):
  mylist = [1,2,3,4]
  print("values inside the function: ",mylist)
  return

mylist = [10,20,30]
change_me(mylist)
print(f"Values outside the function: {mylist}")
```

```python
#contoh sederhana
#lambda function adalah untuk mereplikasi suatu function

#cara penulisan 1 :
def sum(arg1,arg2):
  return arg1+arg2
#cara penulisan lain : 
sum = lambda arg1,arg2 : arg1 + arg2

sum_105 = sum

print(f"Value of total second function : {sum_105(1,5)}")
print(f"Value of total first function  : {sum(2,56)}")
```
